## **🌟 ¿Qué es el aprendizaje supervisado?**

 Es una técnica de aprendizaje automático (machine learning) donde el modelo **aprende a partir de datos etiquetados.** Es decir, cada ejemplo del conjunto de datos tiene una entrada (features) y una salida esperada (etiqueta o label).

 supvsnosup.png


## **🎯 Objetivo**
 El objetivo del modelo supervisado es **predecir la salida** (etiqueta) correcta para nuevas entradas que nunca ha visto, basándose en los patrones aprendidos durante el entrenamiento.

## **🧠 ¿Cómo funciona?**

 Generalmente el modelo se entrena con una porción de los datos disponibles, debemos tener claro que entre mejor calidad de datos tengamos, el modelo responderá de mejor manera y será más preciso.
### En general se consideran tres etapas:
 **Entrenamiento:** se le da al modelo un conjunto de datos con entradas y salidas conocidas.

 **Aprendizaje:** el modelo ajusta sus parámetros internos para minimizar el error entre lo que predice y la etiqueta real.

 **Predicción:** se evalúa el modelo usando nuevos datos y se mide qué tan bien generaliza.

# **🔀 Tipos de problemas en el aprendizaje supervisado**

### Podemos dividir los problemas supervisados en dos grandes categorías:

 **🔹 Clasificación**
 Cuando la salida esperada es una categoría o clase. Por ejemplo, predecir si un correo es "spam" o "no spam", o si una imagen representa un "gato" o un "perro".

 **🔹 Regresión**
 Cuando la salida esperada es un valor numérico continuo. Por ejemplo, predecir el precio de una casa o la temperatura de mañana.

## **🔍 Modelos de Clasificación Supervisada**

En esta sección veremos algunos de los algoritmos más utilizados para clasificación, destacando cómo funcionan, sus ventajas y desventajas, y cuándo podrían fallar.



### **🌳 Árbol de Decisión**

Un árbol de decisión clasifica los datos dividiendo el espacio de entrada en regiones, según condiciones sobre las características. Se construye un modelo en forma de árbol, donde cada nodo representa una pregunta sobre una característica, y las hojas corresponden a una predicción final.

**¿Cómo funciona?**
- Divide los datos de forma recursiva según la característica que mejor separa las clases (usando criterios como *gini* o *entropía*).
- Cada división busca maximizar la "pureza" de los grupos resultantes.
- Una vez que no se pueden hacer más divisiones significativas, se asigna una clase.

**Ventajas**
- Muy fácil de interpretar y visualizar.
- No requiere escalado de variables.
- Puede manejar variables categóricas y numéricas.

**Desventajas**
- Puede sobreajustarse (overfitting), especialmente si es muy profundo.
- Pequeños cambios en los datos pueden generar árboles muy distintos.
- Menor rendimiento comparado con modelos ensamblados.

### 🌲 **Random Forest**

Random forest es un algoritmo de ensamble, es decir, es un método que combina múltiples modelos simples (llamados modelos base) para construir un modelo más robusto y preciso.

En este caso, se tienen múltiples árboles de decisión independientes para posteriormente, combinar las predicciones obtenidas.

En clasificación, toma la clase más votada, y en regresión, promedia las predicciones.

**Por ejemplo**:

🔸 En **clasificación**:
Cada árbol da su propia "opinión" (predicción de clase). Random Forest elige **la clase que más veces se repite** entre todos los árboles, es decir, **la clase más votada**.

Ejemplo:

Supongamos que queremos clasificar una flor y entrenamos 5 árboles. Las predicciones son:

- Árbol 1 → "Iris-setosa"  
- Árbol 2 → "Iris-versicolor"  
- Árbol 3 → "Iris-setosa"  
- Árbol 4 → "Iris-setosa"  
- Árbol 5 → "Iris-versicolor"  

🗳 Votación:
- "Iris-setosa": 3 votos  
- "Iris-versicolor": 2 votos

✅ Resultado final: **"Iris-setosa"** (por mayoría de votos)


🔹 En **regresión**:
Cada árbol predice un valor numérico. Random Forest toma el **promedio** de todas las predicciones para obtener el valor final.

Ejemplo:

Queremos predecir el precio de una casa. Los árboles estiman:

- Árbol 1 → \$180.000  
- Árbol 2 → \$200.000  
- Árbol 3 → \$190.000  
- Árbol 4 → \$210.000  
- Árbol 5 → \$185.000  

📊 Promedio:
\[
\frac{180.000 + 200.000 + 190.000 + 210.000 + 185.000}{5} = 193.000
\]

✅ Resultado final: **\$193.000**

---

🎯 *¿Por qué esto mejora el rendimiento?*  
Porque si un árbol se equivoca, los otros pueden "compensar" su error. Esto **reduce la varianza** del modelo, haciendo que generalice mejor en nuevos datos.

# **Consideraciones en Random Forest**


🔁 **1. Bootstrapping (muestreo aleatorio con reemplazo):**  
En lugar de entrenar todos los árboles con el mismo dataset, cada árbol se entrena con una **muestra aleatoria** del dataset original. Esto significa que algunos datos se repiten y otros quedan fuera en cada árbol, lo que genera diversidad.

🎲 **2. Aleatoriedad en las características:**  
Además, en cada división de un árbol, **no se consideran todas las columnas** (features), sino un subconjunto aleatorio. Esto evita que todos los árboles se parezcan demasiado.

🗳️ **3. Votación o promedio:**  
Una vez entrenados, cada árbol da su predicción:
- En **clasificación**, cada árbol "vota" por una clase y gana la mayoría.
- En **regresión**, se calcula el promedio de todas las predicciones.


### ✅ Ventajas de Random Forest

- **Menos overfitting:** Al combinar muchos árboles distintos, se suavizan los errores individuales.
- **Robusto:** Soporta datos ruidosos y outliers mejor que un solo árbol.
- **Importancia de variables:** Puede decirnos qué variables fueron más útiles para tomar decisiones.
- **Maneja bien datos faltantes:** Algunos árboles pueden usar filas que otros descartan, así no se pierde toda la información.


### ⚠️ Desventajas de Random Forest

- **Difícil de interpretar:** Un árbol individual es fácil de entender, pero cientos de árboles son como una caja negra.
- **Más lento en predicción:** Para hacer una predicción, hay que consultar todos los árboles.
- **Consume más memoria:** Almacenar muchos árboles requiere más recursos que un modelo simple.


# 👥 k-Nearest Neighbors (KNN)

**k-NN** (*k-Nearest Neighbors*) es un algoritmo de aprendizaje supervisado que puede ser utilizado tanto para **clasificación** como para **regresión**. Su funcionamiento se basa en el principio de **similitud**, es decir, asume que los datos que se parecen entre sí (según cierta métrica de distancia) probablemente pertenezcan a la misma clase o tengan valores de salida similares.

A diferencia de otros algoritmos, **k-NN no construye un modelo explícito durante el entrenamiento**. En su lugar, **almacena el dataset completo** y, al momento de realizar una predicción, busca los *k* ejemplos más cercanos al nuevo punto de entrada para tomar una decisión.

- En **clasificación**, asigna la clase más común entre sus *k* vecinos más cercanos.
- En **regresión**, retorna el promedio (o mediana) del valor numérico de sus vecinos.

Su simplicidad lo hace útil como primer modelo base, pero su rendimiento depende fuertemente de:
- La elección adecuada de *k*.
- La métrica de distancia usada (como Euclidiana o Manhattan).
- La escala de las variables (requiere normalización).
- La presencia de ruido o variables irrelevantes.


### 🧠 ¿Cómo funciona?

1. Se define un número **k** (cantidad de vecinos a considerar).
2. Cuando se quiere clasificar un nuevo punto:
   - Se calcula la **distancia** entre ese punto y todos los del dataset (usualmente con distancia Euclidiana).
   - Se seleccionan los **k vecinos más cercanos**.
   - Se asigna al nuevo punto la **clase más común entre esos vecinos**.






image.png

### ✅ Ventajas de KNN

- **Muy fácil de entender e implementar.**
- **No hace suposiciones** sobre la forma de los datos (puede capturar relaciones no lineales).
- **Se adapta bien a nuevos datos**, ya que solo necesita consultar a sus vecinos.

### ⚠️ Desventajas de KNN

- **Lento en predicción** si el dataset es grande, porque tiene que calcular la distancia con todos los puntos.
- **Sensible a la escala de las variables**: si una característica está en una escala mucho mayor que otra, dominará el cálculo de distancia. Por eso es fundamental **normalizar o estandarizar** los datos antes de usar KNN.
- **No funciona bien con muchas variables irrelevantes** (maldición de la dimensionalidad): demasiadas dimensiones pueden hacer que todos los puntos parezcan igualmente lejanos.
- **No se puede interpretar fácilmente** por qué clasificó un punto como lo hizo (solo se basa en proximidad).

### 🎯 ¿Cuándo usar KNN?

- Cuando tienes un dataset **no tan grande** y quieres un enfoque simple.
- Cuando sospechas que la relación entre variables y clases **no es lineal**.
- Cuando puedes **normalizar bien** los datos y reducir las variables irrelevantes.

💡 *Tip útil:*  
En general, se elige *k* con valores impares (como 3, 5, 7) para evitar empates. También se puede ajustar con validación cruzada.

## 📏 Métricas para evaluar modelos de clasificación

Una vez que entrenamos un modelo, necesitamos **medir qué tan bien está funcionando**. No basta con ver si acierta muchas veces: a veces, un modelo puede tener una precisión alta pero estar fallando de forma importante en ciertas clases (como detectar enfermedades o fraudes).


### ✅ Accuracy (exactitud)

Es la proporción de **predicciones correctas** sobre el total de casos.

$$
\text{Accuracy} = \frac{\text{Predicciones correctas}}{\text{Total de casos}}
$$

**Ejemplo:**
- Si tenemos 100 ejemplos y acertamos en 90 → **Accuracy = 90%**
- ❗ Pero ojo: si una clase representa el 95% del dataset, predecir siempre esa clase da alta accuracy… aunque no estés detectando nada útil.


### 🎯 Precision (precisión)

De todos los casos que el modelo **predijo como positivos**, ¿cuántos **realmente lo eran**?

$$
\text{Precisión} = \frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos} + \text{Falsos Positivos}}
$$

Ideal cuando:
- Queremos **minimizar falsos positivos**.
- Ejemplo: detección de spam. No queremos que correos válidos terminen en la carpeta de spam.

### 🔍 Recall (sensibilidad)

De todos los casos que **realmente eran positivos**, ¿cuántos logré predecir correctamente?

$$
\text{Recall} = \frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos} + \text{Falsos Negativos}}
$$

Ideal cuando:
- Queremos **detectar todos los positivos**, aunque fallemos con algunos negativos.
- Ejemplo: diagnóstico médico. Es mejor tener falsos positivos que dejar pasar un caso real.


### ⚖️ F1-Score

Es la **media armónica** entre precisión y recall. Sirve como medida balanceada, especialmente con clases desbalanceadas.

$$
F1 = 2 \cdot \frac{\text{Precisión} \cdot \text{Recall}}{\text{Precisión} + \text{Recall}}
$$

### 📊 Matriz de Confusión

| **Real \ Predicción** | **Positivo**         | **Negativo**         |
|------------------------|----------------------|----------------------|
| **Positivo**           | ✅ Verdadero Positivo (VP) | ❌ Falso Negativo (FN) |
| **Negativo**           | ❌ Falso Positivo (FP)     | ✅ Verdadero Negativo (VN) |


- **VP (Verdadero Positivo)**: el modelo predijo positivo y era positivo.
- **VN (Verdadero Negativo)**: el modelo predijo negativo y era negativo.
- **FP (Falso Positivo)**: el modelo predijo positivo pero era negativo.
- **FN (Falso Negativo)**: el modelo predijo negativo pero era positivo.


# 📊 Matriz de Confusión

Es una tabla que muestra cuántas veces el modelo predijo correctamente o se equivocó con cada clase:

|                        | Predicho Positivo | Predicho Negativo |
|------------------------|-------------------|-------------------|
| **Real Positivo**      | Verdadero Positivo (TP) | Falso Negativo (FN) |
| **Real Negativo**      | Falso Positivo (FP)     | Verdadero Negativo (TN) |

Esto permite **analizar los errores en detalle**.


🎯 **Importante:**
- **Accuracy puede ser engañosa** si hay clases desbalanceadas.
- Siempre conviene mirar varias métricas y **la matriz de confusión** para entender bien cómo está funcionando el modelo.
- Si los resultados no son buenos, **volver a revisar los datos es fundamental**: clases desbalanceadas, variables irrelevantes, falta de normalización, errores de codificación, etc.


# 🌸 Comparación de modelos supervisados con el dataset Iris

En esta sección vamos a aplicar lo aprendido sobre **modelos de clasificación supervisada** usando un conjunto de datos muy conocido en ciencia de datos: el **dataset Iris**.

### 📌 ¿Qué es el dataset Iris?

El dataset Iris contiene **150 observaciones de flores** de tres especies diferentes de la flor *Iris*:
- *Iris-setosa*
- *Iris-versicolor*
- *Iris-virginica*

Cada flor está descrita por **cuatro características numéricas**:
- Largo del sépalo
- Ancho del sépalo
- Largo del pétalo
- Ancho del pétalo

Este dataset es muy útil porque:
- Es balanceado (50 ejemplos por clase)
- Es pequeño y fácil de visualizar
- Permite practicar clasificación multiclase

### 🎯 ¿Qué haremos?

1. **Cargaremos y exploraremos** el dataset.
2. **Entrenaremos dos modelos de clasificación**:
   - 👥 k-Nearest Neighbors (KNN)
   - 🌲 Random Forest
3. **Evaluaremos el desempeño** de ambos modelos usando métricas como:
   - Accuracy
   - Precision
   - Recall
   - F1-Score
   - Matriz de confusión
4. **Compararemos los resultados** para entender las ventajas y desventajas de cada algoritmo.

Este ejercicio nos ayudará a ver **cómo distintos algoritmos pueden comportarse frente al mismo problema**, y por qué no basta con entrenar un modelo: también es clave **evaluar, comparar y reflexionar** sobre los resultados.

```python
# 📥 Cargar y preparar el dataset Iris

# Importamos el dataset de iris desde sklearn
from sklearn.datasets import load_iris
import pandas as pd

# Cargamos el dataset
iris = load_iris()

# Creamos un DataFrame con los datos de las flores (sin etiquetas aún)
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Agregamos una nueva columna llamada "target" que contiene las etiquetas numéricas (0, 1, 2)
df["target"] = iris.target

# Agregamos una columna "species" con los nombres reales de las especies usando los índices
# Esto convierte 0 → 'setosa', 1 → 'versicolor', 2 → 'virginica'
df["species"] = df["target"].apply(lambda x: iris.target_names[x])

# Mostramos las primeras filas del DataFrame para verificar
df.head()

df

```

```python
# Importamos la librería para gráficos
import matplotlib.pyplot as plt

# Creamos un diccionario que asocia cada especie con un color
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

# Definimos el tamaño de la figura del gráfico
plt.figure(figsize=(8, 6))

# Iteramos sobre cada especie única en el DataFrame
for species in df["species"].unique():
    # Filtramos el DataFrame para quedarnos solo con la especie actual
    subset = df[df["species"] == species]

    # Dibujamos un gráfico de dispersión (scatter plot) para esa especie
    plt.scatter(
        subset["sepal length (cm)"],   # Eje x: longitud del sépalo
        subset["sepal width (cm)"],    # Eje y: ancho del sépalo
        label=species,                 # Etiqueta que aparecerá en la leyenda
        color=colors[species]          # Color definido para esa especie
    )

# Añadimos un título al gráfico
plt.title("Distribución de especies según sépalos")

# Etiquetamos los ejes
plt.xlabel("Longitud del sépalo (cm)")
plt.ylabel("Ancho del sépalo (cm)")

# Mostramos la leyenda con los nombres de las especies
plt.legend()

# Activamos una cuadrícula de fondo para mejorar la visualización
plt.grid(True)

# Mostramos el gráfico final
plt.show()
```

```python
import matplotlib.pyplot as plt

# Diccionario de colores para cada especie
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

# Diccionario con nombres traducidos de columnas
columnas_traducidas = {
    "sepal length (cm)": "Longitud del sépalo (cm)",
    "sepal width (cm)": "Ancho del sépalo (cm)",
    "petal length (cm)": "Longitud del pétalo (cm)",
    "petal width (cm)": "Ancho del pétalo (cm)"
}

# Creamos una figura con 2 filas y 2 columnas (4 gráficos)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Lista de combinaciones de columnas a graficar
plots = [
    ("sepal length (cm)", "sepal width (cm)"),
    ("petal length (cm)", "petal width (cm)"),
    ("sepal length (cm)", "petal length (cm)"),
    ("sepal width (cm)", "petal width (cm)")
]

# Títulos para cada gráfico
titles = [
    "Sépalos: longitud vs ancho",
    "Pétalos: longitud vs ancho",
    "Longitud del sépalo vs del pétalo",
    "Ancho del sépalo vs del pétalo"
]

# Iteramos para crear cada subplot
for i, ax in enumerate(axs.flat):
    x_col, y_col = plots[i]

    for species in df["species"].unique():
        subset = df[df["species"] == species]
        ax.scatter(
            subset[x_col],
            subset[y_col],
            label=species,
            color=colors[species]
        )

    # Etiquetas de ejes en español
    ax.set_xlabel(columnas_traducidas[x_col])
    ax.set_ylabel(columnas_traducidas[y_col])
    ax.set_title(titles[i])
    ax.grid(True)

# Agregamos leyenda general abajo
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.05))

# Texto explicativo: lo bajamos un poco más
fig.text(0.5, -0.08, 'Cada punto representa una flor del dataset Iris.',
         ha='center', fontsize=12,
         bbox=dict(facecolor='lightgrey', edgecolor='black'))

# Ajustamos el diseño para que se vea todo bien
plt.tight_layout(rect=[0, 0.1, 1, 0.95])  # rect = (left, bottom, right, top)

# Mostramos el gráfico
plt.show()

```

```python
# Importamos funciones para dividir los datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split
# Importamos el escalador para normalizar los datos (muy importante para KNN)
from sklearn.preprocessing import StandardScaler

# Seleccionamos las variables predictoras (features) usando los nombres de las columnas del dataset iris
X = df[iris.feature_names]

# Seleccionamos la variable objetivo o etiqueta (es decir, la especie de la flor)
y = df["target"]

# Creamos un objeto escalador para normalizar los datos (media = 0, desviación estándar = 1)
scaler = StandardScaler()

# Aplicamos el escalador a los datos de entrada (X)
# Esto es necesario porque KNN es sensible a la escala de los datos
X_scaled = scaler.fit_transform(X)

# Dividimos los datos en conjunto de entrenamiento (70%) y de prueba (30%)
# - stratify=y asegura que cada clase esté representada proporcionalmente
# - random_state=42 asegura que la división sea reproducible, misma aleatoriedad
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y)

```

```python
#Entrenamiento Modelos

# Importamos los modelos supervisados: KNN y Random Forest
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Instanciar modelos
# Creamos una instancia del modelo KNN con 5 vecinos
knn = KNeighborsClassifier(n_neighbors=5)

# Creamos una instancia del modelo Random Forest con semilla para reproducibilidad
rf = RandomForestClassifier(random_state=42)

# Entrenar
# Entrenamos el modelo KNN con los datos de entrenamiento
knn.fit(X_train, y_train)

# Entrenamos el modelo Random Forest con los mismos datos
rf.fit(X_train, y_train)



```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Predicciones: se usan los modelos entrenados (knn y rf) para hacer predicciones con los datos de prueba X_test.
#El resultado es un vector de etiquetas predichas (y_pred_knn y y_pred_rf), que luego se comparan con las verdaderas (y_test).
y_pred_knn = knn.predict(X_test)
y_pred_rf = rf.predict(X_test)


# Reportes
print("📘 KNN - Reporte de clasificación:\n")
print(classification_report(y_test, y_pred_knn, target_names=iris.target_names))

#C.R: genera un resumen con métricas clave como precision, recall, f1-score y accuracy para cada clase.

print("🌲 Random Forest - Reporte de clasificación:\n")
print(classification_report(y_test, y_pred_rf, target_names=iris.target_names))

#support = número de instancias reales de cada clase en el conjunto de prueba.
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np

# 1. Cargar el dataset
diabetes = load_diabetes()
df_diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df_diabetes['target'] = diabetes.target

# 2. Separar variables y target
X = df_diabetes.drop('target', axis=1)
y = df_diabetes['target']

# 3. Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Inicializar modelos
lr = LinearRegression()
rf = RandomForestRegressor(random_state=42)

# 5. Entrenar modelos
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)

# 6. Predicciones
y_pred_lr = lr.predict(X_test)
y_pred_rf = rf.predict(X_test)

# 7. Evaluación
def evaluar_modelo(y_true, y_pred, nombre_modelo):
    print(f"📊 {nombre_modelo}")
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("R² :", r2_score(y_true, y_pred))
    print("")

evaluar_modelo(y_test, y_pred_lr, "Regresión Lineal")
evaluar_modelo(y_test, y_pred_rf, "Random Forest Regressor")
```

### 📏 Métricas de Evaluación en Modelos de Regresión

Para evaluar qué tan bien predice un modelo de regresión, usamos las siguientes métricas:

---

#### 🔹 MAE (Mean Absolute Error)
- Mide el **promedio de los errores absolutos** entre los valores reales y los predichos.
- Nos dice, en promedio, cuánto se equivoca el modelo **sin importar si el error es hacia arriba o hacia abajo**.

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

✅ Fácil de interpretar: si el MAE es 41, significa que el modelo se equivoca en promedio 41 unidades.

---

#### 🔹 MSE (Mean Squared Error)
- Similar al MAE, pero **eleva al cuadrado los errores**.
- Penaliza más fuerte los errores grandes.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

📌 Útil si nos interesa **evitar errores grandes**, pero menos intuitivo de interpretar porque está en **unidades al cuadrado**.

---

#### 🔹 R² (Coeficiente de Determinación)
- Mide qué tanto del comportamiento de los datos **explica el modelo**.
- Varía entre 0 y 1 (a veces puede ser negativo si el modelo es muy malo).

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

- \(R^2 = 1\): predicción perfecta.
- \(R^2 = 0\): el modelo no explica nada.
- 📉 Si es negativo, el modelo es peor que simplemente predecir la media.

---

### 🎯 ¿Qué buscamos?
- Un **MAE bajo** ✅
- Un **MSE bajo** ✅
- Un **R² lo más cercano a 1 posible** ✅


```python
import matplotlib.pyplot as plt

def plot_predicciones(y_true, y_pred, nombre_modelo):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    # Calcular métricas
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    # Gráfico
    plt.figure(figsize=(6, 6))
    plt.scatter(y_true, y_pred, alpha=0.7, color='mediumseagreen', edgecolor='k')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
    plt.xlabel('Valores reales')
    plt.ylabel('Predicciones')
    plt.title(f'{nombre_modelo} - Real vs Predicho')

    # Agregar texto con métricas
    texto = f'MAE: {mae:.2f}\nMSE: {mse:.2f}\nR²: {r2:.2f}'
    plt.text(x=y_true.min(), y=y_true.max()-30, s=texto, fontsize=10,
             bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5'))

    plt.grid(True)
    plt.show()

plot_predicciones(y_test, y_pred_lr, "Regresión Lineal")
plot_predicciones(y_test, y_pred_rf, "Random Forest Regressor")
```
