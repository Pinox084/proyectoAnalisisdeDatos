## **ðŸŒŸ Â¿QuÃ© es el aprendizaje supervisado?**

 Es una tÃ©cnica de aprendizaje automÃ¡tico (machine learning) donde el modelo **aprende a partir de datos etiquetados.** Es decir, cada ejemplo del conjunto de datos tiene una entrada (features) y una salida esperada (etiqueta o label).

 supvsnosup.png


## **ðŸŽ¯ Objetivo**
 El objetivo del modelo supervisado es **predecir la salida** (etiqueta) correcta para nuevas entradas que nunca ha visto, basÃ¡ndose en los patrones aprendidos durante el entrenamiento.

## **ðŸ§  Â¿CÃ³mo funciona?**

 Generalmente el modelo se entrena con una porciÃ³n de los datos disponibles, debemos tener claro que entre mejor calidad de datos tengamos, el modelo responderÃ¡ de mejor manera y serÃ¡ mÃ¡s preciso.
### En general se consideran tres etapas:
 **Entrenamiento:** se le da al modelo un conjunto de datos con entradas y salidas conocidas.

 **Aprendizaje:** el modelo ajusta sus parÃ¡metros internos para minimizar el error entre lo que predice y la etiqueta real.

 **PredicciÃ³n:** se evalÃºa el modelo usando nuevos datos y se mide quÃ© tan bien generaliza.

# **ðŸ”€ Tipos de problemas en el aprendizaje supervisado**

### Podemos dividir los problemas supervisados en dos grandes categorÃ­as:

 **ðŸ”¹ ClasificaciÃ³n**
 Cuando la salida esperada es una categorÃ­a o clase. Por ejemplo, predecir si un correo es "spam" o "no spam", o si una imagen representa un "gato" o un "perro".

 **ðŸ”¹ RegresiÃ³n**
 Cuando la salida esperada es un valor numÃ©rico continuo. Por ejemplo, predecir el precio de una casa o la temperatura de maÃ±ana.

## **ðŸ” Modelos de ClasificaciÃ³n Supervisada**

En esta secciÃ³n veremos algunos de los algoritmos mÃ¡s utilizados para clasificaciÃ³n, destacando cÃ³mo funcionan, sus ventajas y desventajas, y cuÃ¡ndo podrÃ­an fallar.



### **ðŸŒ³ Ãrbol de DecisiÃ³n**

Un Ã¡rbol de decisiÃ³n clasifica los datos dividiendo el espacio de entrada en regiones, segÃºn condiciones sobre las caracterÃ­sticas. Se construye un modelo en forma de Ã¡rbol, donde cada nodo representa una pregunta sobre una caracterÃ­stica, y las hojas corresponden a una predicciÃ³n final.

**Â¿CÃ³mo funciona?**
- Divide los datos de forma recursiva segÃºn la caracterÃ­stica que mejor separa las clases (usando criterios como *gini* o *entropÃ­a*).
- Cada divisiÃ³n busca maximizar la "pureza" de los grupos resultantes.
- Una vez que no se pueden hacer mÃ¡s divisiones significativas, se asigna una clase.

**Ventajas**
- Muy fÃ¡cil de interpretar y visualizar.
- No requiere escalado de variables.
- Puede manejar variables categÃ³ricas y numÃ©ricas.

**Desventajas**
- Puede sobreajustarse (overfitting), especialmente si es muy profundo.
- PequeÃ±os cambios en los datos pueden generar Ã¡rboles muy distintos.
- Menor rendimiento comparado con modelos ensamblados.

### ðŸŒ² **Random Forest**

Random forest es un algoritmo de ensamble, es decir, es un mÃ©todo que combina mÃºltiples modelos simples (llamados modelos base) para construir un modelo mÃ¡s robusto y preciso.

En este caso, se tienen mÃºltiples Ã¡rboles de decisiÃ³n independientes para posteriormente, combinar las predicciones obtenidas.

En clasificaciÃ³n, toma la clase mÃ¡s votada, y en regresiÃ³n, promedia las predicciones.

**Por ejemplo**:

ðŸ”¸ En **clasificaciÃ³n**:
Cada Ã¡rbol da su propia "opiniÃ³n" (predicciÃ³n de clase). Random Forest elige **la clase que mÃ¡s veces se repite** entre todos los Ã¡rboles, es decir, **la clase mÃ¡s votada**.

Ejemplo:

Supongamos que queremos clasificar una flor y entrenamos 5 Ã¡rboles. Las predicciones son:

- Ãrbol 1 â†’ "Iris-setosa"  
- Ãrbol 2 â†’ "Iris-versicolor"  
- Ãrbol 3 â†’ "Iris-setosa"  
- Ãrbol 4 â†’ "Iris-setosa"  
- Ãrbol 5 â†’ "Iris-versicolor"  

ðŸ—³ VotaciÃ³n:
- "Iris-setosa": 3 votos  
- "Iris-versicolor": 2 votos

âœ… Resultado final: **"Iris-setosa"** (por mayorÃ­a de votos)


ðŸ”¹ En **regresiÃ³n**:
Cada Ã¡rbol predice un valor numÃ©rico. Random Forest toma el **promedio** de todas las predicciones para obtener el valor final.

Ejemplo:

Queremos predecir el precio de una casa. Los Ã¡rboles estiman:

- Ãrbol 1 â†’ \$180.000  
- Ãrbol 2 â†’ \$200.000  
- Ãrbol 3 â†’ \$190.000  
- Ãrbol 4 â†’ \$210.000  
- Ãrbol 5 â†’ \$185.000  

ðŸ“Š Promedio:
\[
\frac{180.000 + 200.000 + 190.000 + 210.000 + 185.000}{5} = 193.000
\]

âœ… Resultado final: **\$193.000**

---

ðŸŽ¯ *Â¿Por quÃ© esto mejora el rendimiento?*  
Porque si un Ã¡rbol se equivoca, los otros pueden "compensar" su error. Esto **reduce la varianza** del modelo, haciendo que generalice mejor en nuevos datos.

# **Consideraciones en Random Forest**


ðŸ” **1. Bootstrapping (muestreo aleatorio con reemplazo):**  
En lugar de entrenar todos los Ã¡rboles con el mismo dataset, cada Ã¡rbol se entrena con una **muestra aleatoria** del dataset original. Esto significa que algunos datos se repiten y otros quedan fuera en cada Ã¡rbol, lo que genera diversidad.

ðŸŽ² **2. Aleatoriedad en las caracterÃ­sticas:**  
AdemÃ¡s, en cada divisiÃ³n de un Ã¡rbol, **no se consideran todas las columnas** (features), sino un subconjunto aleatorio. Esto evita que todos los Ã¡rboles se parezcan demasiado.

ðŸ—³ï¸ **3. VotaciÃ³n o promedio:**  
Una vez entrenados, cada Ã¡rbol da su predicciÃ³n:
- En **clasificaciÃ³n**, cada Ã¡rbol "vota" por una clase y gana la mayorÃ­a.
- En **regresiÃ³n**, se calcula el promedio de todas las predicciones.


### âœ… Ventajas de Random Forest

- **Menos overfitting:** Al combinar muchos Ã¡rboles distintos, se suavizan los errores individuales.
- **Robusto:** Soporta datos ruidosos y outliers mejor que un solo Ã¡rbol.
- **Importancia de variables:** Puede decirnos quÃ© variables fueron mÃ¡s Ãºtiles para tomar decisiones.
- **Maneja bien datos faltantes:** Algunos Ã¡rboles pueden usar filas que otros descartan, asÃ­ no se pierde toda la informaciÃ³n.


### âš ï¸ Desventajas de Random Forest

- **DifÃ­cil de interpretar:** Un Ã¡rbol individual es fÃ¡cil de entender, pero cientos de Ã¡rboles son como una caja negra.
- **MÃ¡s lento en predicciÃ³n:** Para hacer una predicciÃ³n, hay que consultar todos los Ã¡rboles.
- **Consume mÃ¡s memoria:** Almacenar muchos Ã¡rboles requiere mÃ¡s recursos que un modelo simple.


# ðŸ‘¥ k-Nearest Neighbors (KNN)

**k-NN** (*k-Nearest Neighbors*) es un algoritmo de aprendizaje supervisado que puede ser utilizado tanto para **clasificaciÃ³n** como para **regresiÃ³n**. Su funcionamiento se basa en el principio de **similitud**, es decir, asume que los datos que se parecen entre sÃ­ (segÃºn cierta mÃ©trica de distancia) probablemente pertenezcan a la misma clase o tengan valores de salida similares.

A diferencia de otros algoritmos, **k-NN no construye un modelo explÃ­cito durante el entrenamiento**. En su lugar, **almacena el dataset completo** y, al momento de realizar una predicciÃ³n, busca los *k* ejemplos mÃ¡s cercanos al nuevo punto de entrada para tomar una decisiÃ³n.

- En **clasificaciÃ³n**, asigna la clase mÃ¡s comÃºn entre sus *k* vecinos mÃ¡s cercanos.
- En **regresiÃ³n**, retorna el promedio (o mediana) del valor numÃ©rico de sus vecinos.

Su simplicidad lo hace Ãºtil como primer modelo base, pero su rendimiento depende fuertemente de:
- La elecciÃ³n adecuada de *k*.
- La mÃ©trica de distancia usada (como Euclidiana o Manhattan).
- La escala de las variables (requiere normalizaciÃ³n).
- La presencia de ruido o variables irrelevantes.


### ðŸ§  Â¿CÃ³mo funciona?

1. Se define un nÃºmero **k** (cantidad de vecinos a considerar).
2. Cuando se quiere clasificar un nuevo punto:
   - Se calcula la **distancia** entre ese punto y todos los del dataset (usualmente con distancia Euclidiana).
   - Se seleccionan los **k vecinos mÃ¡s cercanos**.
   - Se asigna al nuevo punto la **clase mÃ¡s comÃºn entre esos vecinos**.

### âœ… Ventajas de KNN

- **Muy fÃ¡cil de entender e implementar.**
- **No hace suposiciones** sobre la forma de los datos (puede capturar relaciones no lineales).
- **Se adapta bien a nuevos datos**, ya que solo necesita consultar a sus vecinos.

### âš ï¸ Desventajas de KNN

- **Lento en predicciÃ³n** si el dataset es grande, porque tiene que calcular la distancia con todos los puntos.
- **Sensible a la escala de las variables**: si una caracterÃ­stica estÃ¡ en una escala mucho mayor que otra, dominarÃ¡ el cÃ¡lculo de distancia. Por eso es fundamental **normalizar o estandarizar** los datos antes de usar KNN.
- **No funciona bien con muchas variables irrelevantes** (maldiciÃ³n de la dimensionalidad): demasiadas dimensiones pueden hacer que todos los puntos parezcan igualmente lejanos.
- **No se puede interpretar fÃ¡cilmente** por quÃ© clasificÃ³ un punto como lo hizo (solo se basa en proximidad).

### ðŸŽ¯ Â¿CuÃ¡ndo usar KNN?

- Cuando tienes un dataset **no tan grande** y quieres un enfoque simple.
- Cuando sospechas que la relaciÃ³n entre variables y clases **no es lineal**.
- Cuando puedes **normalizar bien** los datos y reducir las variables irrelevantes.

ðŸ’¡ *Tip Ãºtil:*  
En general, se elige *k* con valores impares (como 3, 5, 7) para evitar empates. TambiÃ©n se puede ajustar con validaciÃ³n cruzada.

## ðŸ“ MÃ©tricas para evaluar modelos de clasificaciÃ³n

Una vez que entrenamos un modelo, necesitamos **medir quÃ© tan bien estÃ¡ funcionando**. No basta con ver si acierta muchas veces: a veces, un modelo puede tener una precisiÃ³n alta pero estar fallando de forma importante en ciertas clases (como detectar enfermedades o fraudes).


### âœ… Accuracy (exactitud)

Es la proporciÃ³n de **predicciones correctas** sobre el total de casos.

$$
\text{Accuracy} = \frac{\text{Predicciones correctas}}{\text{Total de casos}}
$$

**Ejemplo:**
- Si tenemos 100 ejemplos y acertamos en 90 â†’ **Accuracy = 90%**
- â— Pero ojo: si una clase representa el 95% del dataset, predecir siempre esa clase da alta accuracyâ€¦ aunque no estÃ©s detectando nada Ãºtil.


### ðŸŽ¯ Precision (precisiÃ³n)

De todos los casos que el modelo **predijo como positivos**, Â¿cuÃ¡ntos **realmente lo eran**?

$$
\text{PrecisiÃ³n} = \frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos} + \text{Falsos Positivos}}
$$

Ideal cuando:
- Queremos **minimizar falsos positivos**.
- Ejemplo: detecciÃ³n de spam. No queremos que correos vÃ¡lidos terminen en la carpeta de spam.

### ðŸ” Recall (sensibilidad)

De todos los casos que **realmente eran positivos**, Â¿cuÃ¡ntos logrÃ© predecir correctamente?

$$
\text{Recall} = \frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos} + \text{Falsos Negativos}}
$$

Ideal cuando:
- Queremos **detectar todos los positivos**, aunque fallemos con algunos negativos.
- Ejemplo: diagnÃ³stico mÃ©dico. Es mejor tener falsos positivos que dejar pasar un caso real.


### âš–ï¸ F1-Score

Es la **media armÃ³nica** entre precisiÃ³n y recall. Sirve como medida balanceada, especialmente con clases desbalanceadas.

$$
F1 = 2 \cdot \frac{\text{PrecisiÃ³n} \cdot \text{Recall}}{\text{PrecisiÃ³n} + \text{Recall}}
$$

### ðŸ“Š Matriz de ConfusiÃ³n

| **Real \ PredicciÃ³n** | **Positivo**         | **Negativo**         |
|------------------------|----------------------|----------------------|
| **Positivo**           | âœ… Verdadero Positivo (VP) | âŒ Falso Negativo (FN) |
| **Negativo**           | âŒ Falso Positivo (FP)     | âœ… Verdadero Negativo (VN) |


- **VP (Verdadero Positivo)**: el modelo predijo positivo y era positivo.
- **VN (Verdadero Negativo)**: el modelo predijo negativo y era negativo.
- **FP (Falso Positivo)**: el modelo predijo positivo pero era negativo.
- **FN (Falso Negativo)**: el modelo predijo negativo pero era positivo.


# ðŸ“Š Matriz de ConfusiÃ³n

Es una tabla que muestra cuÃ¡ntas veces el modelo predijo correctamente o se equivocÃ³ con cada clase:

|                        | Predicho Positivo | Predicho Negativo |
|------------------------|-------------------|-------------------|
| **Real Positivo**      | Verdadero Positivo (TP) | Falso Negativo (FN) |
| **Real Negativo**      | Falso Positivo (FP)     | Verdadero Negativo (TN) |

Esto permite **analizar los errores en detalle**.


ðŸŽ¯ **Importante:**
- **Accuracy puede ser engaÃ±osa** si hay clases desbalanceadas.
- Siempre conviene mirar varias mÃ©tricas y **la matriz de confusiÃ³n** para entender bien cÃ³mo estÃ¡ funcionando el modelo.
- Si los resultados no son buenos, **volver a revisar los datos es fundamental**: clases desbalanceadas, variables irrelevantes, falta de normalizaciÃ³n, errores de codificaciÃ³n, etc.

```python
# ðŸ“¥ Cargar y preparar el dataset Iris

# Importamos el dataset de iris desde sklearn
from sklearn.datasets import load_iris
import pandas as pd

# Cargamos el dataset
iris = load_iris()

# Creamos un DataFrame con los datos de las flores (sin etiquetas aÃºn)
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Agregamos una nueva columna llamada "target" que contiene las etiquetas numÃ©ricas (0, 1, 2)
df["target"] = iris.target

# Agregamos una columna "species" con los nombres reales de las especies usando los Ã­ndices
# Esto convierte 0 â†’ 'setosa', 1 â†’ 'versicolor', 2 â†’ 'virginica'
df["species"] = df["target"].apply(lambda x: iris.target_names[x])

# Mostramos las primeras filas del DataFrame para verificar
df.head()

df

```

```python
# Importamos la librerÃ­a para grÃ¡ficos
import matplotlib.pyplot as plt

# Creamos un diccionario que asocia cada especie con un color
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

# Definimos el tamaÃ±o de la figura del grÃ¡fico
plt.figure(figsize=(8, 6))

# Iteramos sobre cada especie Ãºnica en el DataFrame
for species in df["species"].unique():
    # Filtramos el DataFrame para quedarnos solo con la especie actual
    subset = df[df["species"] == species]

    # Dibujamos un grÃ¡fico de dispersiÃ³n (scatter plot) para esa especie
    plt.scatter(
        subset["sepal length (cm)"],   # Eje x: longitud del sÃ©palo
        subset["sepal width (cm)"],    # Eje y: ancho del sÃ©palo
        label=species,                 # Etiqueta que aparecerÃ¡ en la leyenda
        color=colors[species]          # Color definido para esa especie
    )

# AÃ±adimos un tÃ­tulo al grÃ¡fico
plt.title("DistribuciÃ³n de especies segÃºn sÃ©palos")

# Etiquetamos los ejes
plt.xlabel("Longitud del sÃ©palo (cm)")
plt.ylabel("Ancho del sÃ©palo (cm)")

# Mostramos la leyenda con los nombres de las especies
plt.legend()

# Activamos una cuadrÃ­cula de fondo para mejorar la visualizaciÃ³n
plt.grid(True)

# Mostramos el grÃ¡fico final
plt.show()
```

```python
import matplotlib.pyplot as plt

# Diccionario de colores para cada especie
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

# Diccionario con nombres traducidos de columnas
columnas_traducidas = {
    "sepal length (cm)": "Longitud del sÃ©palo (cm)",
    "sepal width (cm)": "Ancho del sÃ©palo (cm)",
    "petal length (cm)": "Longitud del pÃ©talo (cm)",
    "petal width (cm)": "Ancho del pÃ©talo (cm)"
}

# Creamos una figura con 2 filas y 2 columnas (4 grÃ¡ficos)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Lista de combinaciones de columnas a graficar
plots = [
    ("sepal length (cm)", "sepal width (cm)"),
    ("petal length (cm)", "petal width (cm)"),
    ("sepal length (cm)", "petal length (cm)"),
    ("sepal width (cm)", "petal width (cm)")
]

# TÃ­tulos para cada grÃ¡fico
titles = [
    "SÃ©palos: longitud vs ancho",
    "PÃ©talos: longitud vs ancho",
    "Longitud del sÃ©palo vs del pÃ©talo",
    "Ancho del sÃ©palo vs del pÃ©talo"
]

# Iteramos para crear cada subplot
for i, ax in enumerate(axs.flat):
    x_col, y_col = plots[i]

    for species in df["species"].unique():
        subset = df[df["species"] == species]
        ax.scatter(
            subset[x_col],
            subset[y_col],
            label=species,
            color=colors[species]
        )

    # Etiquetas de ejes en espaÃ±ol
    ax.set_xlabel(columnas_traducidas[x_col])
    ax.set_ylabel(columnas_traducidas[y_col])
    ax.set_title(titles[i])
    ax.grid(True)

# Agregamos leyenda general abajo
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.05))

# Texto explicativo: lo bajamos un poco mÃ¡s
fig.text(0.5, -0.08, 'Cada punto representa una flor del dataset Iris.',
         ha='center', fontsize=12,
         bbox=dict(facecolor='lightgrey', edgecolor='black'))

# Ajustamos el diseÃ±o para que se vea todo bien
plt.tight_layout(rect=[0, 0.1, 1, 0.95])  # rect = (left, bottom, right, top)

# Mostramos el grÃ¡fico
plt.show()

```

```python
# Importamos funciones para dividir los datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split
# Importamos el escalador para normalizar los datos (muy importante para KNN)
from sklearn.preprocessing import StandardScaler

# Seleccionamos las variables predictoras (features) usando los nombres de las columnas del dataset iris
X = df[iris.feature_names]

# Seleccionamos la variable objetivo o etiqueta (es decir, la especie de la flor)
y = df["target"]

# Creamos un objeto escalador para normalizar los datos (media = 0, desviaciÃ³n estÃ¡ndar = 1)
scaler = StandardScaler()

# Aplicamos el escalador a los datos de entrada (X)
# Esto es necesario porque KNN es sensible a la escala de los datos
X_scaled = scaler.fit_transform(X)

# Dividimos los datos en conjunto de entrenamiento (70%) y de prueba (30%)
# - stratify=y asegura que cada clase estÃ© representada proporcionalmente
# - random_state=42 asegura que la divisiÃ³n sea reproducible, misma aleatoriedad
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y)

```

```python
#Entrenamiento Modelos

# Importamos los modelos supervisados: KNN y Random Forest
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Instanciar modelos
# Creamos una instancia del modelo KNN con 5 vecinos
knn = KNeighborsClassifier(n_neighbors=5)

# Creamos una instancia del modelo Random Forest con semilla para reproducibilidad
rf = RandomForestClassifier(random_state=42)

# Entrenar
# Entrenamos el modelo KNN con los datos de entrenamiento
knn.fit(X_train, y_train)

# Entrenamos el modelo Random Forest con los mismos datos
rf.fit(X_train, y_train)



```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Predicciones: se usan los modelos entrenados (knn y rf) para hacer predicciones con los datos de prueba X_test.
#El resultado es un vector de etiquetas predichas (y_pred_knn y y_pred_rf), que luego se comparan con las verdaderas (y_test).
y_pred_knn = knn.predict(X_test)
y_pred_rf = rf.predict(X_test)


# Reportes
print("ðŸ“˜ KNN - Reporte de clasificaciÃ³n:\n")
print(classification_report(y_test, y_pred_knn, target_names=iris.target_names))

#C.R: genera un resumen con mÃ©tricas clave como precision, recall, f1-score y accuracy para cada clase.

print("ðŸŒ² Random Forest - Reporte de clasificaciÃ³n:\n")
print(classification_report(y_test, y_pred_rf, target_names=iris.target_names))

#support = nÃºmero de instancias reales de cada clase en el conjunto de prueba.
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np

# 1. Cargar el dataset
diabetes = load_diabetes()
df_diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df_diabetes['target'] = diabetes.target

# 2. Separar variables y target
X = df_diabetes.drop('target', axis=1)
y = df_diabetes['target']

# 3. Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Inicializar modelos
lr = LinearRegression()
rf = RandomForestRegressor(random_state=42)

# 5. Entrenar modelos
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)

# 6. Predicciones
y_pred_lr = lr.predict(X_test)
y_pred_rf = rf.predict(X_test)

# 7. EvaluaciÃ³n
def evaluar_modelo(y_true, y_pred, nombre_modelo):
    print(f"ðŸ“Š {nombre_modelo}")
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("RÂ² :", r2_score(y_true, y_pred))
    print("")

evaluar_modelo(y_test, y_pred_lr, "RegresiÃ³n Lineal")
evaluar_modelo(y_test, y_pred_rf, "Random Forest Regressor")
```

### ðŸ“ MÃ©tricas de EvaluaciÃ³n en Modelos de RegresiÃ³n

Para evaluar quÃ© tan bien predice un modelo de regresiÃ³n, usamos las siguientes mÃ©tricas:

---

#### ðŸ”¹ MAE (Mean Absolute Error)
- Mide el **promedio de los errores absolutos** entre los valores reales y los predichos.
- Nos dice, en promedio, cuÃ¡nto se equivoca el modelo **sin importar si el error es hacia arriba o hacia abajo**.

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

âœ… FÃ¡cil de interpretar: si el MAE es 41, significa que el modelo se equivoca en promedio 41 unidades.

---

#### ðŸ”¹ MSE (Mean Squared Error)
- Similar al MAE, pero **eleva al cuadrado los errores**.
- Penaliza mÃ¡s fuerte los errores grandes.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

ðŸ“Œ Ãštil si nos interesa **evitar errores grandes**, pero menos intuitivo de interpretar porque estÃ¡ en **unidades al cuadrado**.

---

#### ðŸ”¹ RÂ² (Coeficiente de DeterminaciÃ³n)
- Mide quÃ© tanto del comportamiento de los datos **explica el modelo**.
- VarÃ­a entre 0 y 1 (a veces puede ser negativo si el modelo es muy malo).

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

- \(R^2 = 1\): predicciÃ³n perfecta.
- \(R^2 = 0\): el modelo no explica nada.
- ðŸ“‰ Si es negativo, el modelo es peor que simplemente predecir la media.

---

### ðŸŽ¯ Â¿QuÃ© buscamos?
- Un **MAE bajo** âœ…
- Un **MSE bajo** âœ…
- Un **RÂ² lo mÃ¡s cercano a 1 posible** âœ…


```python
import matplotlib.pyplot as plt

def plot_predicciones(y_true, y_pred, nombre_modelo):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    # Calcular mÃ©tricas
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    # GrÃ¡fico
    plt.figure(figsize=(6, 6))
    plt.scatter(y_true, y_pred, alpha=0.7, color='mediumseagreen', edgecolor='k')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
    plt.xlabel('Valores reales')
    plt.ylabel('Predicciones')
    plt.title(f'{nombre_modelo} - Real vs Predicho')

    # Agregar texto con mÃ©tricas
    texto = f'MAE: {mae:.2f}\nMSE: {mse:.2f}\nRÂ²: {r2:.2f}'
    plt.text(x=y_true.min(), y=y_true.max()-30, s=texto, fontsize=10,
             bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5'))

    plt.grid(True)
    plt.show()

plot_predicciones(y_test, y_pred_lr, "RegresiÃ³n Lineal")
plot_predicciones(y_test, y_pred_rf, "Random Forest Regressor")
```
